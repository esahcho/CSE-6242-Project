```python
import pandas as pd
```

```python
CITY_REGIONS = {
 'Phoenix': {'state': 'AZ', 'region': 'Desert'},
 'Los Angeles': {'state': 'CA', 'region': 'Pacific Coast'},
 'Atlanta': {'state': 'GA', 'region': 'South'},
 'Chicago': {'state': 'IL', 'region': 'Midwest'},
 'Boston': {'state': 'MA', 'region': 'Northeast'},
 'Denver': {'state': 'CO', 'region': 'Southwest'},
 'Seattle': {'state': 'WA', 'region': 'Pacific Coast'}
}
```

```python
df = pd.read_parquet("gradient_boosting_forecasts_by_city.parquet")
```

```python
def map_city_to_region(df, city_col="city"):
    # create mapping dictionaries -adding state and regions
    city_to_state = {city: info["state"] for city, info in CITY_REGIONS.items()}
    city_to_region = {city: info["region"] for city, info in CITY_REGIONS.items()}

    # map to dataframe
    df["state"] = df[city_col].map(city_to_state)
    df["region"] = df[city_col].map(city_to_region)

    return df
```

```python
new_forecast = map_city_to_region(df, city_col="city")
new_forecast.to_parquet("gradient_boost_final.parquet", index=True)
#print(new_forecast.head())
```
